# Distribution of Pairing Scores for Each R > 1

## r = 2

```python
%matplotlib inline
from pymongo import MongoClient
import numpy as np
import matplotlib.pyplot as plt

client = MongoClient()
db = client['yummly']
combinations = db['combinations1k']

r = 2

pipeline = [
    {
        "$match": {
            "r": r
        }
    },
    {
        "$project": {
            "score": 1
        }
    }
]

cursor = combinations.aggregate(pipeline)

# http://matplotlib.org/1.2.1/examples/pylab_examples/histogram_demo.html
#import matplotlib.mlab as mlab

# mu, sigma = 100, 15
x = np.array([c['score'] for c in cursor if c['score'] < 0.2])
mu = x.mean()

# the histogram of the data
#n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green', alpha=0.75)
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green')
plt.axvline(x=mu)

# add a 'best fit' line
#y = mlab.normpdf( bins, mu, sigma)
#l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Scores')
plt.ylabel('Probability')
plt.title(r'Histogram of Pairing Scores Given r=2')
#plt.axis([40, 160, 0, 0.03])
#plt.grid(True)

plt.show()
```

## r = 3

```python
r = 3

pipeline = [
    {
        "$match": {
            "r": r
        }
    },
    {
        "$project": {
            "score": 1
        }
    }
]

cursor = combinations.aggregate(pipeline)
x = np.array([c['score'] for c in cursor if c['score'] < 0.2])

n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green')
plt.axvline(x=mu)
plt.title(r'Histogram of Pairing Scores Given r=3')
plt.show()
```
